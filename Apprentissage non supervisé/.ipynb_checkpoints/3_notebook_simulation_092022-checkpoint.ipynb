```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
from sklearn.preprocessing import StandardScaler
```

```python
# Custom colors for charts

c_0 = np.array([2, 48, 71,256])/256
c_1 = np.array([251, 133, 0,256])/256
c_2 = np.array([255, 183, 3,256])/256
c_3 = np.array([33, 158, 188,256])/256
c_4 = np.array([142, 202, 230,256])/256

# Creating a custom colormap
# The code comes from https://matplotlib.org/3.1.1/tutorials/colors/colormap-manipulation.html
from matplotlib.colors import ListedColormap

N = 256
vals = np.ones((N, 4))
vals[:, 0] = np.linspace(c_0[0], c_1[0], N)
vals[:, 1] = np.linspace(c_0[1], c_1[1], N)
vals[:, 2] = np.linspace(c_0[2], c_1[2], N)
custom_cmp1 = ListedColormap(vals)

# A colormap with white in the middle
N = 256
vals = np.ones((N, 4))
vals[:, 0] = np.concatenate((np.linspace(c_0[0], 1, 128), np.linspace(1, c_1[0], 128)), axis = None)
vals[:, 1] = np.concatenate((np.linspace(c_0[1], 1, 128), np.linspace(1, c_1[1], 128)), axis = None)
vals[:, 2] = np.concatenate((np.linspace(c_0[2], 1, 128), np.linspace(1, c_1[2], 128)), axis = None)
custom_cmp2 = ListedColormap(vals)
```

**Aggregation and Importing the data**

```python
df_customers = pd.read_csv('olist_customers_dataset.csv')
df_orders = pd.read_csv('olist_orders_dataset.csv')
df_payments = pd.read_csv('olist_order_payments_dataset.csv')
df_reviews = pd.read_csv('olist_order_reviews_dataset.csv')

# Adding all orders with unique customer id
df_full = df_customers.loc[:,['customer_id', 'customer_unique_id']]

# Adding order value, review score and time
df_full = pd.merge(df_full,
                   
                   pd.merge(df_orders.loc[:,['customer_id', 'order_id', 'order_purchase_timestamp']],
                            
                            pd.merge(df_payments.loc[:,['order_id','payment_value']],
                                     df_reviews.loc[:,['order_id','review_score']],
                                     
                                     on = 'order_id'),
                            
                            on = 'order_id'),
                  
                  on = 'customer_id')

# Convert to datetime
df_full['order_purchase_timestamp'] = pd.to_datetime(df_full['order_purchase_timestamp'])


# Unwated fields are eliminated
df_full = df_full.drop(['customer_id'], axis = 1)
```

```python
df_full
```

```python
def aggregate_data(df):
    # For aggregating the data from a dataframe
    
    # Recency
    ## Set the most recent as d0
    most_recent = df['order_purchase_timestamp'].max()

    ## Calculate the number of days from the most recent
    df.loc[:,'recency'] = (most_recent - df['order_purchase_timestamp']).dt.days # .dt.days convert datime to nb of days
    
    df = df.groupby('customer_unique_id').agg(frequency = ('order_id', 'count'),
                                              monetary_value = ('payment_value', 'sum'),
                                              recency = ('recency', 'min'),
                                              review_mean_score = ('review_score', 'mean'))
    df['frequency'] = np.log1p(df['frequency'])
    df['monetary_value'] = np.log1p(df['monetary_value'])
    
    return df
    

def simmulation(df_full, start_date, step, update_frequency, end_date = None):
    
    # Date at which start the simulation, data before are considered in the root model
    start_date = pd.to_datetime(start_date)
    
    # Data at which the simulation stops
    if end_date == None:
        end_date = df_full['order_purchase_timestamp'].max()
    else:
        end_date = pd.to_datetime(end_date)
    
    spent_time = 0 # Time spent between two updates
    
    results = []
    
    
    # databases
    ## database before adding new samples
    df_root = aggregate_data(df_full[df_full['order_purchase_timestamp'] < start_date])
    scaler_root = StandardScaler().fit(df_root)
    df_root_sc = scaler_root.transform(df_root)
    clt_root = KMeans(n_clusters = 4).fit(df_root_sc)
    
    for t in pd.date_range(start = start_date, end = (end_date - pd.DateOffset(step)), freq = f'{step}D'): #Date will be iterated comparing t and t
        ## Keep info of when update have been performed
        updated = bool(False)
        
        ## database containg the new samples
        df_updated = aggregate_data(df_full[df_full['order_purchase_timestamp'] < t])
        
        ## updating the root database if necessary
        if spent_time > update_frequency:
            df_root = df_updated
            scaler_root = StandardScaler().fit(df_root)
            df_root_sc =scaler_root.transform(df_root)
            clt_root = KMeans(n_clusters = 4).fit(df_root_sc)

            spent_time = 0
            updated = True
        else:
            spent_time += step
        
        # data preprocessing
        df_updated_sc = StandardScaler().fit_transform(df_updated)
        
        # models
        ## root model
        labels_pred = clt_root.predict(scaler_root.transform(df_updated))
        
        ## updated model
        clt_updated = KMeans(n_clusters = 4).fit(df_updated_sc)
        labels_updated = clt_updated.labels_
        
        # Claculate rand index
        rand_score = adjusted_rand_score(labels_updated, labels_pred)
        
        # append results
        results.append([t.to_pydatetime(), rand_score, updated])
        
    return np.array(results)
```

**With no maintenance**

```python
simmulation_0 = simmulation(df_full, '2017-07-01', 5, 10000, None)
```

```python
plt.plot(simmulation_0[:,0],
         simmulation_0[:,1],
         color = c_0,
         linestyle = '--')

plt.scatter(simmulation_0[simmulation_0[:,2]==True][:,0],
            simmulation_0[simmulation_0[:,2]==True][:,1],
            color = c_1,
            marker = 'v')

plt.xticks(rotation=90)
plt.ylabel('ARI')

plt.ylim(0,1.1)
```
